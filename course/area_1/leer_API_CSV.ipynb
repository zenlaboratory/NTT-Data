```python
'''
Escribe una función que, dado un número indeterminado de números, devuelva dos cosas: 
- 1. La suma de los mismos. 
- 2. El valor mayor. Recuerda escribirlo de dos formas, con y sin return.   '''

def numeros(*nums):
    suma = 0
    for numero in nums:
        suma += numero
    
    numero_Mayor = 0
    for i in nums:
        if i > numero_Mayor:
            numero_Mayor = i
    mensaje = print(f'La suma de los números {nums} es: {suma} \ny el mayor es: {numero_Mayor}')
    return mensaje

numeros(1, 5, 14, 9, 18)
    
```

```python
import urllib
import json

http_request = urllib.request.urlopen("https://data.sfgov.org/resource/qz8r-q5ne.json")
 
#Leemos los datos de la url.
datos = http_request.read()
 
#conversión a json
datos_json = json.loads(datos)
# datos_json
```

```python
suma = 0
conteo = 0

for elemento in datos_json:
    if 'average_speed_of_answer_in_secs' in elemento:
        suma += float(elemento['average_speed_of_answer_in_secs'])
        conteo += 1

media = suma / conteo

print(f'La media es: {media:.2f}')
```

```python
''' API:
USO. https://data.sfgov.org/resource/qz8r-q5ne.json '''

import pandas as pd

df_data = pd.read_json('https://data.sfgov.org/resource/qz8r-q5ne.json')
df_data
```

```python
import csv

suma = 0
conteo = 0
 
with open('Call_Center_Metrics_for_the_Health_Service_System_20240513.csv', 'r') as archivo_csv:
    # Crea un objeto lector CSV. Genera diccionario por linea
    lector_csv = csv.DictReader(archivo_csv)

    for fila in lector_csv:
        if fila['Average Speed of Answer in Secs'].strip():
            suma += float(fila['Average Speed of Answer in Secs'])
            conteo += 1

media = suma / conteo

print(f'La media es: {media:.2f}')
```
